70. Climbing Stairs

# [leetcode](https://leetcode.com/problems/climbing-stairs/)

You are climbing a staircase. It takes n steps to reach the top.

Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top?

# Reasoning

[neetcodevideo](https://www.youtube.com/watch?v=Y0lT9Fck7qI)

This is a `dynamic programming` problem.  

Each time we have two choices: take one or two steps.  
A better way to visualize thep roblem is to consider it as a `decision tree` where at every node we hat two splits. At the end we need to count the number of leafs.  
This decision tree is a `recursion`, and requires reqursive algorithm, either a `DFS` or `BFS`.  
The time complexity of a `DFS` solution is 2 * Height of the tree ~ O(2^n).  

__NOTE__: there are parts of the decision tree that are _repeated_. If we have the same node, than the paths would be identical. So we can _store_ result of one in `DP`, a cash and use it later. 

`!` This cashing is called `momoization`

`!` DFS can be implemented with cashing the result.  

In the end, we arrive to the DFS where we need to compute __a single passage__ as we cover all the node values from 0 to n. Then, we can re-use the results. This is O(n) time complexity.  

### True dynamic programming solution

__NOTE__ each node split depends on all others. So we can _start at the bottom_ and work our way upwards. This is `DP-Bottom UP` approach.  

We start with the base case. The result is stored in array `DP` the size n.  
At the base case, at n we have only one solution.  
```python
DP = [ ..., ..., ..., 1 ]
```

Next we compute the next to the last solution, and from 4 to 5 we can only take one step, so, one way
```python
DP = [ ..., ..., 1, 1 ]
```

__NOTE__: these two steps are independent of the totan 'n'

Next we consider the 2nd from the last position and see how many ways can we reach the result. It depends on the results from the previos steps, Thus we only need to see the number of ways to reach the prevous resutls. 
We _sum_ the results from two previos iterations.  

```python
DP = [ ..., 1+1, 1, 1 ]
```

For the previos position it is the same way  
```python
DP = [ 2+1, 1+1, 1, 1 ]
```

In the end, the final answer would be the one we one we get the last.  
This is very similar to the __Fibbonacy number problem__.  

Using the DP array requires us to have O(n) extra memroy. However, it is technically not needed, as we only use the last two values. So we can overwrite them all the time or shift the


Then, the time complexity is O(n-1) and space complexity is O(1)


````python
class Solution:
    def climbStairs(self, n: int) -> int:
        one = 1 # steps to reach with 1
        two = 1 # steps to reach with 2
        for i in range (n-1):
            temp = one
            one = one + two # add two perevious values 
            two = temp 
        return one

            
````
