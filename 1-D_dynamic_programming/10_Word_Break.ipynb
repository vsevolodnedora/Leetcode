# 139. Word Break

[leetcode](https://leetcode.com/problems/word-break/description/)

Given a string s and a dictionary of strings wordDict, return true if s can be segmented into a space-separated sequence of one or more dictionary words.

Note that the same word in the dictionary may be reused multiple times in the segmentation.

# Reasoning 

[neetcodevideo]

Bruteforce solution has two approaches.  
First, we can try it char by char, whether each of them is in the dict. This is a two pointer solution, where we start the right untill we find a match, if we do not, we start from the second etc... 

Checking every prefix, every word, then we would need to do it O(n*n) times.  

Second approach is to chck every single word in the dict and see if it can be found in the string. Then we mask the found word in the string and repeat the analysis. 

This solution is O(n*m) complexity.  

This is more efficient solution. 

However, checking if a word is in the dict is also has a comlexity so the overall time complexity is O(n*m*n).  

### Decision tree solution

Start at i=0, the starting point that either 0 if we do not found a substring or a i=len(word). The decision tree has a brach for each word in the dict, whether it has it or not. If there is a word, we continue the branch. If not, we do not ontinue. Then, at the next level, we considr all remaing words and _shift the pointer_ to the next word. Once we find a word, we repeat. 

Here we can use casch to simplfy the solution. We casch the failed branches.   

Once we are at the last index, we can return the True. 

For each index in dp[i] we will have True and False depending on whether we can find the solution. 

### Bottom-up approach. 

Starting from the last character, we fill the dp[i] array wih eithet True or False depending on whether the string[i:len(str)-1] can be found in the dictionary. 



```python
from typing import List
# does not work for some reason
class Solution:
    def wordBreak(self, s: str, wordDict: List[str]) -> bool:
        # make a hasch
        dp = [False] * (len(s) + 1) # 1 is fr base case (out of bounds)
        dp[len(s)] = True # once we are at the end of the string, we can return True
        # loop through all values
        for i in range(len(s)-1,-1,-1): # from he back to the front
            for w in wordDict: # check if word matches this portion
                # first check the length
                if ((i+len(w)) <= len(s)  # there is enough characters for comparison
                    and s[i : i+len(w)] == w): # if the word is exactly in the dict
                    dp[i] == dp[i + len(w)] # if we found a single way to break the str starting at i
                # if we found the way to split; we do not need to check other words
                if dp[i]:
                    break
        return dp[0]
    
class Solution:
  def wordBreak(self, s: str, wordDict: List[str]) -> bool:
    word_set = set(wordDict)  # convert wordDict to a set for constant time lookup
    n = len(s)
    dp = [False] * (n+1)  # create an array dp of length n+1
    dp[0] = True  # empty string can be segmented into an empty sequence of words
    
    for i in range(1, n+1):
        for j in range(i):
            if dp[j] and s[j:i] in word_set:
                dp[i] = True
                break
    
    return dp[n]  # return the final answer
```
