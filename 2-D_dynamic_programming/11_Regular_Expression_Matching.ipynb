# 10. Regular Expression Matching

[leetcode](https://leetcode.com/problems/regular-expression-matching/description/)

Given an input string s and a pattern p, implement regular expression matching with support for '.' and '*' where:

'.' Matches any single character.​​​​
'*' Matches zero or more of the preceding element.
The matching should cover the entire input string (not partial).

# Reasoning

[neetcodevideo](https://www.youtube.com/watch?v=HAA8mgxlov8)

Note, s='aa' and p ='a' do not match  
While s='aa' and p='a*' can match as * can be any mynber of 'a's; so True  
Consider s='ab', p='.*'; here '.' is ''.   

### Bruteforce solution.  

Consider a decision tree for s='aa' and p = 'a*'; 
We have a choice; either to repeat 'a' once, or 0 times. 
If we do not repeat the char, than the decision tree stops, as there are no other characters.  

If we do repeat a for a '*' than we move down and now we have a choice whether to again repeat it once or not. If we do, we get 'aa' and it is what we need. So we recovered the solution.  

A decision tree, however, is not very efficient, as we have to repeat making a decision 2^n times.  
However, this is a DP problemt, i.e., there is repeated work that can be cahed.  
We can do here a Top-Down DP with cahe and a Bottom-Up (within cache).  

Here we consider a Top-Down DP solution.  

Consdier s='aab' and p='c*a*b*'.  
Let's set two ponters for each string 'i' and 'j'.  
How, if we see a '*' we have choices whether we repeat a character or not. 
__NOTE__ a '*' also means that there may be 0 characters. So, if we use '*' in the first position, we remove 'c' from the 'p' string. We also shift a pointer to 'a' now.  

Thus, we make a decision tree following 'nothing' branch and pointers became (i,j+2), 
whre +2 means that we also moved by a star.  
Next we consider 'a*' and we make the choice, whether to add 1+ 'a' or just set 0 'a'.  
We follow both braches. However, we know that we need 'a' so perhaps we can skip it here.  

Following chosing 'a' with out '*' we now have 'ca' as a second string and pointers chaged to (i+1,j). Note, the 'j' is constant as we just removed star.  

Next we have a and '*' still so we can ether repeat 'a' again or just skip it and move to 'b'. 
If we add another 'a', so we get 'aa' -- this matches the string 'aab' (the first part). 
Indexes change as (i+1, j)  

Next, we again still have a star, and we can either use it or not. If we do, we get 'aaa' which is not what we need. So we chose not to use this '*' this time, so we get to 'aab'.

Finally, we got a string 'aab' that is what we wanted to get.  

To check if we are perfectly matched we check if both indeces got out of bounds.  

Note, only one of the strings can go out of bounds.  

If j is out of boudns we must return false.  

```python
class Solution:
    def isMatch(self, s: str, p: str) -> bool:
        """ top=-down memoization solution """
        cache = {} # 
        # recursive DFS
        def dfs(i,j):
            """ i,j, - positions in the input strings """
            # check if already computed
            if (i,j) in cache:
                return cache[(i,j)]
            # base case 
            if (i>=len(s) and j>=len(p)):
                # we found a solution
                return True
            if (j>=len(p)):
                # we cannot have a match
                return False
            # check if there is a match between first char 
            match = (i < len(s)) and (s[i]==p[j] or p[j] == ".") # as "." matchs any char
            # if match # first char matches exactly
            # Handle : check of "*" as it has the highest precidene
            if ((j+1)<len(p) and p[j+1] == "*"):
                # we need to take care of a start via recurisive call
                #opt1 = dfs(i,j+2) # Option 1: do not use the star
                #opt2 = dfs(i+1,j) and match # Option 2: use the star (but only of there is match) 
                # if any of them is true
                cache[(i,j)] = dfs(i,j+2) or (dfs(i+1,j) and match)
                # if (opt1 or opt2):
                    # return cache[(i,j)]
                return cache[(i,j)]

            # Handle if there is no '*' 
            if match:
                cache[(i,j)] = dfs(i+1,j+1)
                return cache[(i,j)]
            
            cache[(i,j)] = False

            return False
        
        return dfs(0,0)
```
