# 309. Best Time to Buy and Sell Stock with Cooldown

[leetcode](https://leetcode.com/problems/best-time-to-buy-and-sell-stock-with-cooldown/)

You are given an array prices where prices[i] is the price of a given stock on the ith day.

Find the maximum profit you can achieve. You may complete as many transactions as you like (i.e., buy one and sell one share of the stock multiple times) with the following restrictions:

After you sell your stock, you cannot buy stock on the next day (i.e., cooldown one day).
Note: You may not engage in multiple transactions simultaneously (i.e., you must sell the stock before you buy again).

# Reasoning

[neetcodevideo](https://www.youtube.com/watch?v=I7j0F7AHpb8&t=5s)

The complicated part about this problem is to make choices when to uy and when to sell.  
A drawing shelps.  

Consider prices = [1,2,3,0,2]

At index 1 we has a choice - buy or sell. 
Here we can only buy as we do not have it. but 
we can do cooldown. 

The decision tree at index 0 has three branches Buy Cooldown Sell, 
but sell requires us to have the stock.  
- Buy - negative cost (loss)
- Cooldown - 0 cost
- Sell - positive cost (gain)

After one branch, we repeat the tree building.  

Note, -- after buying/selling we must take a cooldown.  

Every leaf node of the tree contains the final decisions and gives the maximum.  

The time complexity of this solution is related to the size of the decision tree, which is O(2^n).  

The dynamic probramming solution, using `hashing` allows to obtain O(n) solution.  

The key for hash is [i, buy/sell] where the latter is a boolean.  

The size of the hash is O(n) and the time complexity is O(n)

```python
from typing import List
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        # state Buy/sell
        # if buy -> i + 1 (index for cooldown tracking)
        # if sell -> i + 2
        dp = {} # key = (i, buying) val = max_profit 
        # consider recursive func
        def dfs(i, buying):
            # base case 
            if (i >= len(prices)):
                return 0 # no profit
            # if the pair is already computed
            if (i, buying) in dp:
                return dp[(i,buying)]
            # decision (depends only on the state)
            if buying:
                # substract price from the total and get the ma value for 
                # the remaining array
                buy = dfs(i+1, not buying) - prices[i]# get the profit if buying
                cooldown = dfs(i+1, buying)
                dp[(i,buying)] = max(buy,cooldown)# get max profit
            else:
                # we must add a coodown day (so +2)
                sell = dfs(i+2, not buying) + prices[i]
                cooldown = dfs(i+1, buying)
                dp[(i,buying)] = max(sell,cooldown)# get max profit
            return dp[(i,buying)]
        return dfs(0,True)
```
