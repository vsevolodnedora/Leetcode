# 1143. Longest Common Subsequence

[leetcode](https://leetcode.com/problems/longest-common-subsequence/description/)

Given two strings text1 and text2, return the length of their longest common subsequence. If there is no common subsequence, return 0.

A subsequence of a string is a new string generated from the original string with some characters (can be none) deleted without changing the relative order of the remaining characters.

For example, "ace" is a subsequence of "abcde".
A common subsequence of two strings is a subsequence that is common to both strings.

# Reasoning

__NOTE__ 'absde' and 'ace' have 3 common subsequence lenth, as we can remove values from the first.  

This is 2D DP program. 

__NOTE__ we can see a subproblem here. Once we find that characters are the same, we can break it and consider the rest of it. If they are not the same, we can remove a char from either of the string and consider the remainder. 

A 2D DP problem can be visualized with grid:

  [ a c e ]
a [ x . . ] 0
b [ . x . ] 0
c [ . x . ] 0
d [ . . x ] 0
e [ . . x ] 0
    0 0 0   0

To fill this grid, we need to start from the end. 
If characters are the same, we go diagonally.  
If they are not, we go down or up. 

Once we are ut of bounds, we add values to 0

After we finihed the pass, we go back a d add ones if chars match

A 2D DP problem can be visualized with grid:

  [ a c e ]
a [ 3 . . ] 0
b [ . 2 . ] 0
c [ . 2 . ] 0
d [ . . 1 ] 0
e [ . . 1 ] 0
    0 0 0   0

With code we will try to solve it "brute-force" ish. 
We need to compute all values for all entries in the grid.  
SOlution via `bottom-up DP`. 

Given i - col and j - row

Each values in the cell is filled based on whether characters match.  



```python
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        dp = [[0 for j in range(len(text2)+1)] for i in range(len(text1)+1)] # grid
        for i in range(len(text1)-1,-1,-1): # start at the bottom right
            for j in range(len(text2)-1,-1,-1):
                if text1[i] == text2[j]:
                    dp[i][j] = 1 + dp[i+1][j+1] # add diagonal and 1 for match
                else:
                    dp[i][j] = max(dp[i][j+1],dp[i+1][j]) # right abd below and their max
        return dp[0][0]
```
