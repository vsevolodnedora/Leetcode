# 97. Interleaving String

[leetcode](https://leetcode.com/problems/interleaving-string/description/)

Given strings s1, s2, and s3, find whether s3 is formed by an interleaving of s1 and s2.

An interleaving of two strings s and t is a configuration where s and t are divided into n and m 
substrings
 respectively, such that:

s = s1 + s2 + ... + sn
t = t1 + t2 + ... + tm
|n - m| <= 1
The interleaving is s1 + t1 + s2 + t2 + s3 + t3 + ... or t1 + s1 + t2 + s2 + t3 + s3 + ...
Note: a + b is the concatenation of strings a and b.

# Reasoning

[neetcodevideo](https://www.youtube.com/watch?v=3Rw3p9LrgvE&ab_channel=NeetCode)

Given three strings and we need to form the third by interleaving the other two. E.g., can we split them into substrings. Note, the relative order of substrings from a given string is important. In other words, we cannot reshuffle them.  

So the first constraint, is the the number of characters must match.  
The second, -- relative order of earch of the strings must be maintained.  

the optimal solution, with true DP can be given with O(n*m) complexity, where n and m are lengths of two strings.  

### Consider a decision tree. 

Given two strings s1 = aabcc and s2 = abbca, we initialize two pointers i1 and i2; the i3 pointer is in the string 3.  
__NOTE__ i1 + i2 = i3

If s1[i1] == s3[i3]; we i1++ (we found that character from string s1 correspond to the character in s3). 

if both s1[i1] == s2[i2] == s3[i3]; then we have to make decision. At this point our decision tree branches out. Note, that we haubproblems here. 
We can cash the result and then the memory complexity. O(n*m) (memoization solution)

Consider thrue DP slution.  
Two given strings can be represented as follows:  
Add a base case to the table, as if two characterss do form a target string.  
Each cell in the grid will have True or False. Out of bounce posiiton is also valid as length of strings may be different. Only when both strings do not have a character at the end then there is T
_ _ 0 1 2 3 4 .  
_ _ d b b c a .  
0 a   
1 a  
2 b   T  
3 c  
4 c  
. .   

Example 2 & 2 will be 'b' and 'b'; check third string at i=4; it is 'b' so we put True

```python
class Solution:
    def isInterleave(self, s1: str, s2: str, s3: str) -> bool:
        """ mmoization recursive solution """
        if len(s1) + len(s2) != len(s3):
            return False
        dp = {} # cash
        # k = i + j
        def dfs(i,j): 
            """ dfs """
            # base case : out of bounds
            if (i==len(s1) and j==len(s2)):
                return True
            # base case : already computed
            if (i,j) in dp:
                return dp[(i,j)]
            # Recursive case for an _in bounds string_ 
            # does the char matches the string s3[i+j] char 
            # increment the corresponding pointer in the first string
            if (i<len(s1) and s1[i]==s3[i+j] and dfs(i+1,j)):
                return True
            # do the same for the string 2
            if (j<len(s2) and s2[j]==s3[i+j] and dfs(i,j+1)):
                return True
            # if neither returns true, we cash False and return false
            dp[(i,j)] = False
            return False
        return dfs(0,0)
    def isInterleave(self, s1: str, s2: str, s3: str) -> bool:
        """ DP solution """
        if len(s1) + len(s2) != len(s3):
            return False
        # init cash size of n * m and for out of bounds we set
        dp = [ [False] * (len(s2)+1) for i in range(len(s1)+1) ]
        # corner values are true 
        dp[len(s1)][len(s2)] = True
        # start at the bottom and work our way up
        for i in range(len(s1),-1,-1):
            for j in range(len(s2),-1,-1):
                #  if i is in bounds; check charecters; if true; check dp
                if (i<len(s1) and s1[i]==s3[i+j] and dp[i+1][j]):
                    dp[i][j] = True
                # do the same for the string 2
                if (j<len(s2) and s2[j]==s3[i+j] and dp[i][j+1]):
                    dp[i][j] = True
                # the default False is already initialized
        return dp[0,0]
```
