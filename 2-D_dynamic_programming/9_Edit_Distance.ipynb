# 72. Edit Distance

[leetcode](https://leetcode.com/problems/edit-distance/description/)

Given two strings word1 and word2, return the minimum number of operations required to convert word1 to word2.

You have the following three operations permitted on a word:

Insert a character
Delete a character
Replace a character

# Reasoning

[neetcodevideo](https://www.youtube.com/watch?v=XYi2-LPrwm4&ab_channel=NeetCode)

If two strings are identical, we can return 0; as it takes 0 operations to convert onto another. 
If word1 = abc and word2 = "" we can call _delete_ three times are convert 1 into 2. The N operation = lenth of string 1.  
If w1 = "" and w2 = "abs" we can call insert three times and convert w1 into w2.  

Note, it is always possible to convert one into another. 

Consider 2 pointers for two strings and consider them character by character. If chars are equal we increment both pointers. If they are not equal, we can use one of the three operations. Consider all three of them. If we _insert_a character, firtst pointer gets moved and a match is automatically accomplished as we insert the required char.

Replace a character from w1 in w2; we force characters to match -- we shift both pointers

If we delete a cahracter, we remove the unmatched char and shift a pointer; if there is not match again, we keep pointers 

Summary: 
- Insert: (i, j+1)
- delete: (i+1; j)
- replace: (i+1, j+1)

Consider a grid 

  a c d 
a .   3
b  x  2
d 3 2 1
       0
In each position we store a minimum number of operation to make them equal. 
Where 'x' we need t brute force it and find a solution

We fill this table bottom-up. Then find a minimum value for up, down and diagonal. 



 

```python
class Solution:
    def minDistance(self, word1: str, word2: str) -> int:
        cache = [ [float("inf")]*(len(word2)+1) for i in range(len(word1)+1)] # extend for base cases
        # initialize base cases
        # fill bottom row
        for j in range(len(word2)+1):
            cache[len(word1)][j] = len(word2) - j
        # fill last column
        for i in range(len(word1)+1):
            cache[i][len(word2)] = len(word1) - i
        # Go from bottom
        for i in range(len(word1)-1,-1,-1):
            for j in range(len(word2)-1,-1,-1):
                # cases:
                if (word1[i]==word2[j]):
                    # good place; cache
                    min_num_of_op = cache[i+1][j+1]
                    cache[i][j] = min_num_of_op
                else:
                    op_del = cache[i+1][j]
                    op_ins = cache[i][j+1]
                    op_rep = cache[i+1][j+1]
                    min_num_of_op = 1 + min(op_del,op_ins,op_rep) # add 1 as there is at least one operation
                    cache[i][j] = min_num_of_op
        return cache[0][0]
        

```
